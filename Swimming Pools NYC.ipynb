### Matching pool names from NYC Parks data to names in the outdoor sessions and aquatics programmming data

```python
# First list
list1 = {
    'Flushing Meadows Corona Park Aquatics Center',
    'Offsite Locations', 'Constance Baker Motley Recreation Center',
    'Roy Wilkins Recreation Center', "St. John's Recreation Center",
    'Asser Levy Recreation Center', 'Astoria Pool',
    'Chelsea Recreation Center', 'Crotona Pool',
    'Metropolitan Pool & Recreation Center',
    'Gertrude Ederle Recreation Center',
    "St. Mary's Recreation Center", 'Red Hook Pool', 'Wagner College',
    'Lyons Pool', 'Hamilton Fish Pool', 'Liberty Pool', 'Fisher Pool'
}

# Second list
list2 = {
    'Lindower Park Mini Pool', 'Faber Pool', 'Sheltering Arms Pool',
    'General Douglas MacArthur Mini Pool', 'PS 14 Playground Mini Pool',
    'Marcus Garvey Pool', 'Mapes Pool', 'Claremont Pool', 'Floating Pool',
    'John Jay Pool', 'Highbridge Pool', 'West Brighton Pool',
    'Commodore Barry Pool', 'Haffen Pool', 'Hamilton Fish Pool',
    'Sunset Park Pool', 'Bushwick Pool', 'Thomas Jefferson Pool',
    'Asser Levy Pool', 'Liberty Pool', 'Foster Pool', 'Astoria Pool',
    'Crotona Pool', 'Lyons Pool', 'Fisher Pool', 'PS 186 Playground Mini Pool',
    'Marie Curie Park Mini Pool', 'McCarren Park Pool', 'Kosciuszko Pool',
    'Windmuller Mini Pool', 'Douglas and DeGraw Pool', 'Tottenville Pool',
    'Red Hook Pool', 'Jackie Robinson Pool', 'Howard Pool',
    'Van Cortlandt Pool', 'PS 20 Playground Mini Pool',
    'JHS 57 / HS 26 Mini Pool', 'PS 46 Playground Mini Pool',
    'David Fox / PS 251 Mini Pool', 'Wagner Pool', 'Betsy Head Pool',
    'Abe Lincoln Mini Pool', 'Frederick Douglass Mini Pool',
    'Vesuvio Mini Pool', 'Tompkins Square Mini Pool', 'Totten Pool',
    'Dry Dock Pool', 'Thompson Mini Pool', 'Glenwood Houses Mini Pool',
    'Edenwald Houses Mini Pool'
}

# Get unique pool names from your table (assuming it's in a DataFrame `df`)
grouped_pools = set(locations['NAME'])  # Replace with your actual column name if different

# Intersections
match1 = grouped_pools & list1
match2 = grouped_pools & list2

# Optionally, show unmatched:
unmatched1 = list1 - grouped_pools
unmatched2 = list2 - grouped_pools

print("Pools in group and list 1:", sorted(match1))
print("Pools in group and list 2:", sorted(match2))
print("List 1 items not in group:", sorted(unmatched1))
print("List 2 items not in group:", sorted(unmatched2))
```

```python
import numpy as np
```

```python
from rapidfuzz import process, fuzz

# Combine both lists into one set of candidate pool names
all_pool_names = set(list1).union(list2)

# Define a matching function
def get_best_match(pool_name, choices, score_cutoff=90):
    match, score, _ = process.extractOne(pool_name, choices, scorer=fuzz.ratio)
    return match if score >= score_cutoff else np.nan
```

```python
# Apply fuzzy matching to create 'locations' column
locations['matched_pool'] = locations['NAME'].apply(lambda x: get_best_match(x, all_pool_names))
```

## Below is scratchwork

```python
from rapidfuzz import process
```

```python
# Create a mapping from the lessons dataset to best match in the attendance dataset
group1_names = lessons['Swimming Pool'].unique()
group2_names = df['Pool'].unique()

mapping = {
    name: process.extractOne(name, group2_names)[0]
    for name in group1_names
}

# Apply mapping to harmonize
lessons['Swimming Pool Matched'] = lessons['Swimming Pool'].replace(mapping)
```

```python
from rapidfuzz import process, fuzz
```

```python
# We'll store the best match and its score in a list
fuzzy_matches = []

# Convert target list to avoid repeated conversion
lessons_pool_list = lessons['Swimming Pool Matched'].dropna().unique().tolist()

# Iterate through each name in hvi_pools
for pool_name in hvi_pools['matched_pool']:
    if pd.notna(pool_name):
        match, score, _ = process.extractOne(
            pool_name,
            lessons_pool_list,
            scorer=fuzz.token_sort_ratio
        )
        fuzzy_matches.append((pool_name, match, score))
    else:
        fuzzy_matches.append((pool_name, None, 0))
```

```python
# Convert to DataFrame for easier inspection
fuzzy_df = pd.DataFrame(fuzzy_matches, columns=['matched_pool', 'best_match', 'score'])
```

```python
# Merge with hvi_pools for context
hvi_pools_fuzzy = hvi_pools.merge(fuzzy_df, on='matched_pool')
```

```python
strong_matches = hvi_pools_fuzzy[hvi_pools_fuzzy['score'] >= 90]
```

```python
strong_matches.shape
```

```python
strong_matches.head()
```

```python
lessons_hvi = lessons.merge(strong_matches, right_on='best_match', left_on='Swimming Pool Matched', how='left')
```
